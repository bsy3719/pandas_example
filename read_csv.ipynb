```python
path = "./data/titanic.csv"
delimiter = ","
column_type = {'PassengerId': 'int'}
num_row = 20
skip_row = 5
skip_footer = 10
encoding = "utf-8"
use_column = ["PassengerId", "Survived"]
order_column = ['Survived', 'PassengerId']
rename_column = {'PassengerId': 'Id'}
orient = "records"
```

```python
def load_file(path, sep, usecols, dtype, nrows, encoding):
    return pd.read_csv(path, sep=sep, usecols=usecols, dtype=dtype, nrows=nrows, encoding=encoding)
```

```python
def get_preview(path, delimiter, column_type, num_row, skip_row, skip_footer, encoding, use_column, order_column,
            rename_column, orient):

    df = load_file(path, delimiter, use_column, column_type, num_row, encoding)
    
    df = df[order_column]
    df = df.rename(columns=rename_column)
    #df = update_column_order(df, order_column)
    #df = rename_column(df, rename_column)

    df_json = df.to_json(orient=orient)
    result = {"columnData" : df_json}
    result.update({"columns" : {"col1" : 1, "col2" : 2}})

    return result
```
