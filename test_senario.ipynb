```python
class UserDefinedRunFunction():
    #     OUT_REPLACE = "replace"
    #     OUT_CONCAT = "concat"
    #     OUT_NONE = "none"

    def one_hot_encoding(df, col: str):
        """
        name : 원 핫 인코딩
        outputType : concat
        comment : 원핫인코딩 함수, 꼭 1개의 인자를 넣어주세요
        conditionStatus : false
        """
        return pd.get_dummies(df[col])

    def drop(df, index):
        """
        name : 행 삭제
        outputType : none
        comment : 행 삭제 함수
        conditionStatus : true
        """
        return df.drop(index)
    
    def fillNaMedian(df, targetCol:int, groupByCol:list):
        """
        name : 빈값을 중간값으로 채우는 함수
        outputType : replace
        comment : 대상컬럼과 그룹핑할 컬럼을 넣어주세요
        conditionStatus : false
        """
        return df[targetCol].fillna(df.groupby(groupByCol)[targetCol].transform("median"))
    
    def binningAge(df, col):
        """
        name : 컬럼 비닝
        outputType : replace
        comment : Age컬럼을 입력해주세요.
        conditionStatus : false
        """
        df = df.loc[(df["Age"] < 9) | (df["Age"] > 60)]
        bins = [9, 16, 25, 35, 60]
        labels = [1,2,3,4]
        return pd.cut(df[col], bins=bins, labels=labels)
    
    
    # def dropna(df, cols: list):
    #     # df.dropna(subset = ["Sex", "Cabin"], thresh=2400, how='any')
    #     """
    #     name : 결측치 제거
    #     outputType : none
    #     comment : 결측치 제거 함수, 꼭 1개의 인자를 넣어주세요
    #     """
    #     return df.dropna(subset=cols, axis=1)

class UserDefinedApplyFunction():

    def add(df, col1: int, col2: int):
        """
        name : 컬럼 더하기
        outputType : replace
        comment : 두개의 인자 값을 더해주는 함수! 꼭 2개의 인자를 넣어주세요.
        conditionStatus : false
        """
        return df[col1] + df[col2]

    # columns을 던질때 복수형이면서 int형은 어떻게 처리할 것 인지
    def substr(df, col: int, num: int):
        """
        name : 문자 슬라이싱
        outputType : replace
        comment : 문자를 자르는 함수 꼭 1개의 인자를 넣어주세요.
        conditionStatus : false
        """
        return df[col][:-int(num)]
    
    def extractName(df, col:str):
        """
        name : 이름에서 값을 추출하는 함수
        outputType : none
        comment : 반드시 Name 컬럼을 넣어주세요
        conditionStatus : false
        """
        if "Mr" in df[col]:
            return "Mr"
        elif "Miss" in df[col]:
            return "Miss"
        elif "Mrs" in df[col]:
            return "Mrs"
        else:
            return "etc"
```
