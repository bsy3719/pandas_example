```python
def to_parquet(df, path, **kwargs):
    params = {'compression':'gzip'}
    params.update(**kwargs)
    #print(kwargs)
    df.to_parquet(path, **params)
```

```python
def read_parquet(path, **kwargs):
    params = {}
    params.update(**kwargs)
    #print(kwargs)
    df = pd.read_parquet(path, **params)
    return df
```
